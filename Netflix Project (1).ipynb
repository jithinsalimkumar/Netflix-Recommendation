###  Visual Analysis - Univariate, Bivariate after pre-processing of the data

###### Comparing movies and TV shows on Netflix

```python
plt.figure(figsize=(5,5))
typ=df["type"].dropna().value_counts().rename_axis('type').reset_index(name='counts')
sns.barplot(data=typ, x="type",y='counts')
plt.title('Comparison of movies and TV shows')
plt.show()
```

###### <b>Insights :</b>Total number of movies is more than TV shows

###### Trend of content produced on Netflix over the last 22 years

```python
plt.figure(figsize=(10,8))
data_movies=df[df['type']=="Movie"]
x_movies=list(data_movies.groupby(['release_year'])['title'].nunique().tail(22).index) 
y_movies=list(data_movies.groupby(['release_year'])['title'].nunique().tail(22))
sns.lineplot(x=x_movies,y=y_movies,label="Movie")
data_shows=df[df['type']=="TV Show"]
x_shows=list(data_shows.groupby(['release_year'])['title'].nunique().tail(22).index) 
y_shows=list(data_shows.groupby(['release_year'])['title'].nunique().tail(22))
sns.lineplot(x=x_shows,y=y_shows,label="TV Show")
plt.xticks(np.arange(2000, 2022, step=2), rotation = 90)
plt.xlabel('Release year') 
plt.ylabel("Release count")
plt.title("Trend of content on Netflix")
plt.legend()
plt.show()
```

###### <b>Insights :</b>Till 2018 the trend of movies was higher than that of TV shows.But after 2018 movies is showing decrease in trend.In 2021 TV shows trend is higher than that of movies. 

###### <b>Recommendation :</b>Netflix should produce more TV shows than movies as per the trend.

###### Release Year of Movies/TV shows from 2000

```python
plt.figure(figsize=(10,8))
ry=df['release_year'].astype("Int64")
gh = sns.histplot(x = ry,bins=np.arange(2000, 2022, step=2))
plt.title('Release Year of movies/TV shows from 2000')
plt.xlabel('Release year')
for i in gh.patches:
    gh.annotate('{:.0f}'.format(i.get_height()),(i.get_x()+2.5,i.get_height()+50) )
plt.xticks(np.arange(2000, 2022, step =2))
plt.show()
```

###### <b>Insights :</b>Most of Movies/TV shows are released between the year 2018 and 2020
